```python
from pprint import pprint
from Bio import SeqIO
from Bio.SeqFeature import SeqFeature, FeatureLocation
import pandas as pd
```

```python
scaffolds_names = {}
features = {}

def discard(s):
  if s[0] == '>' or s[0] == '<': s = s[1:]
  return s

with open("gms2.lst", 'r') as handle:
  for line in handle:
    if line[0] == '#' or len(line.split()) == 0:
      continue
    text = line.split()
    if 'SequenceID' in text[0]:
      key = text[1]
      if key not in features:
        features[key] = {}
    else:
      if '+' in text[1]: strand = 1
      else: strand = -1
      SF = SeqFeature(FeatureLocation(int(discard(text[2])),
                                        int(discard(text[3])),
                                        strand=strand), type="CDS")
      features[key][text[0]] = SF
      SF.qualifiers['locus_tag'] = [text[0]]
      scaffolds_names[text[0]] = key
pprint(features)
```

```python
ccus = {}

with open("scaffolds.hits_from_MIL_1.txt", 'r') as handle:
  for line in handle:
    text = line.split()
    name = text[1].split('_')[2]
    features[scaffolds_names[text[0]]][text[0]].qualifiers['notes'] = name
    ccus[name] = text[0]
```

```python
for gene in SeqIO.read("T_oleivorans_MIL_1.gbk", 'genbank').features:
  if 'protein_id' in gene.qualifiers and gene.qualifiers['protein_id'][0] in ccus:
    number = ccus[gene.qualifiers['protein_id'][0]]
  else: continue
  feature = features[scaffolds_names[number]][number]
  feature.qualifiers['product'] = gene.qualifiers['product']
```

```python
for rec in SeqIO.parse("proteins.fasta", 'fasta'):
  text = rec.description.split()
  features[text[1]][text[0]].qualifiers['translation'] = [rec.seq]
```

```python
entries = []

for entry in SeqIO.parse("out_scaffold.fa", 'fasta'):
  entry.annotations['molecule_type'] = 'DNA'
  entry.annotations['organism'] = 'Thalassolituus oleivorans Barents'
  entry.annotations['source'] = 'Ivan Valentinovich Antonov'
  entry.annotations['date'] = '2021-12-18'
  entry.features = list(features[entry.id].values())
  entry.format('genbank')
  entries.append(entry)
entries
```

```python
SeqIO.write(entries, "Genome.gbk", 'genbank')
```
